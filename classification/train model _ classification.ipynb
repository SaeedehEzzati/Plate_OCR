```python
from ultralytics import YOLO
from datetime import datetime
import torch
```

```python
from ultralytics import YOLO
torch.cuda.empty_cache()
```

```python
model = YOLO("YOLOv8x-cls.pt")  # load a pretrained YOLOv8n model
# # 
```

```python
start = datetime.now()
model.train(data="C:/Users/s.ezati/Desktop/metaX/plate OCR/classification/output", 
   epochs=100,
   imgsz=64,
   name='yolov8x_cl_custom_100')  # train the model
end = datetime.now()
print(end - start)
```

```python
start = datetime.now()

model.val()  # evaluate model performance on the validation set

end = datetime.now()
print(end - start)
```
