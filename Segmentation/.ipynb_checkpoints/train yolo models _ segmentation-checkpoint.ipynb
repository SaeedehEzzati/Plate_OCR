```python
from ultralytics import YOLO
from datetime import datetime
```

```python
model = YOLO("yolov5x.pt")  # load a pretrained YOLOv8n model
```

```python
start = datetime.now()
model.train(data="data.yaml", 
   epochs=100,
   batch=4,
   name='yolov5x_custom')  # train the model
end = datetime.now()
print(end - start)
```
